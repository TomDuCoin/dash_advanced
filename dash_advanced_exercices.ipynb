### Task 4
Maintenant vous allez réaliser les bollinger bands de chaque courbe sur la 3eme page: 
<br>
https://www.fidelity.com/learning-center/trading-investing/technical-analysis/technical-indicator-guide/bollinger-bands



<br>
<br>
Pour obtenir la 'upper band', pour chaque jour vous ferez ce calcul:
<br>

(((valeur_actuel - moyenne(20 derniers jours)) * 2) + valeur_actuel)
<br>

Même chose pour la 'lower band':
<br>

(((valeur_actuel - moyenne(20 derniers jours)) * 2) - valeur_actuel)
<br>

Vous remplirez l'espace entre les deux courbes sur le graph grâce à ce lien :):<br>
https://plotly.com/python/filled-area-plots/

Et enfin vous transformerez les courbes que vous avez créé à la task 3 par des candlesticks:
https://plotly.com/python/candlestick-charts/
